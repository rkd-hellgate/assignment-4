**VISUALIZATION OF THE AIRBNB DATASET**

<u>COUNTPLOT OF NUMBER OF LISTINGS BY NEIGHBOURHOOD GROUP</u>


```python
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.countplot(x='neighbourhood_group', data=df1, palette='viridis')
plt.title('Number of Listings by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Count')
plt.show()
```

*INFERENCE : FROM THE ABOVE VISUALIZATIZATION WE CAN SEE THAT "MANHATTAN" HAVE THE HIGHEST COUNT OF THE "NEIGHBOURHOOD GROUP" WHILE "STATEN ISLAND" HAVE THE LOWEST COUNT IN THE DATASET* 

<u>PIE CHART FOR DISTRIBUTION OF ROOM TYPES</u>


```python
room_type_counts = df1['room_type'].value_counts()
light_palette = plt.cm.get_cmap('Pastel2', len(room_type_counts))
plt.figure(figsize=(5,5))
plt.pie(room_type_counts, labels=room_type_counts.index, autopct='%1.1f%%', startangle=140, colors=light_palette.colors)
plt.title('Distribution of Room Types')
plt.show()
```

*INFERENCE : FROM THE ABOVE VIZUALIZATION WE CAN SAY THAT MAJORITY OF THE AIRBNB HAVE ENTIRE HOME/APT LISTED WITH (55%) WHILE PRIVATE ROOM AND SHARED ROOM FOLLOWS WITH (44.2%) AND (0.8%)*

<u>HISTPLOT OF DISTRIBUTION OF AIRBNB PRICES</u>


```python
plt.figure(figsize=(10, 6))
sns.histplot(df1['price'], bins=30, kde=True, color='blue')
plt.title('Distribution of Airbnb Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

*INFERENCE : FROM THE ABOVE HISTOGRAM WE CONCLUDED THAT ITS HIGHLY SKEWED WHERE MAJORITY PRICES LIES IN RANGE OF WITHIN 100$ WHILE SOME PRICES MAY VARY IN SOME SCENERIOS*  

<u>BOXPLOT OF DISTRIBUTION OF MINIMUM NIGHTS</u>


```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='minimum_nights', data=df1, color='green')
plt.title('Distribution of Minimum Nights')
plt.xlabel('Minimum Nights')
plt.show()
```

*INFERENCE : FROM THE ABOVE BOXPLOT WE CAN SEE MAJORITY OF PEOPLE BOOK THE AIRBNB FOR 1,2,3 NIGHTS WHILE THERE MAYBE SOME OUTLIERS IN OUR DATASET WHERE THEY HAVE MINIMUM NIGHTS IN AIRBNB TO BE 100 OR 200 WHICH MAYBE UNTRUE MEANING THE DATA MAYBE FALSE*

<u>SCATTERPLOT BETWEEN NUMBER OF REVIEWS VS REVIEWS PER MONTH</u>


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='number_of_reviews', y='reviews_per_month', data=df1, hue='neighbourhood_group', palette='viridis')
plt.title('Number of Reviews vs. Reviews per Month')
plt.xlabel('Number of Reviews')
plt.ylabel('Reviews per Month')
plt.show()
```

*INFERENCE :THIS SCATTER PLOT CHECKS THE RELATIONSHIP BETWEEN THE NUMBER OF REVIEWS AND THE AVERAGE NUMBER OF REVIEWS PER MONTH. IT HELPS TO VISUALIZE WHETHER PROPERTIES WITH MORE REVIEWS ALSO TEND TO HAVE A HIGHER AVERAGE REVIEW FREQUENCY PER MONTH.*

<u>LINEPLOT BETWEEN PRICE VS AVAILABILITY</u>


```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df1, x='availability_365', y='price', color='purple')
plt.title('Price vs. Availability')
plt.xlabel('Availability (in days)')
plt.ylabel('Price')
plt.show()
```

*INFERENCE : THIS TIME SERIES PLOT CHECKS THE RELATIONSHIP BETWEEN AVAILABILITY AND PRICE OVER THE COURSE OF THE YEAR. IT HELPS TO VISUALIZE IF THERE ARE ANY TRENDS OR PATTERNS IN THE AVAILABILITY OF LISTINGS THROUGHOUT THE YEAR AND HOW THEY CORRELATE WITH PRICES.*

**VISUALIZATION OF HR DATASET**

<u>PAIR PLOT OF ENGAGEMENT SURVEY,ABSENCES AND SPECIAL PROJECTS COUNT</u>


```python
sns.pairplot(df2[['EngagementSurvey', 'Absences', 'SpecialProjectsCount']], height=3)
plt.suptitle('Pairplot of Engagement Survey, Absences, and Special Projects Count', y=1.02)
plt.show()
```

* INFERENCE : There seems to be a weak negative correlation between EngagementSurvey scores and Absences, indicating that employees with higher engagement tend to have fewer absences.
There might be a slight positive correlation between EngagementSurvey scores and SpecialProjectsCount, suggesting that more engaged employees might be involved in more special projects.

<u>VIOLENPLOT OF ABSENCES DISTRIBUTION BY DEPARTMENT AND PERFORMANCE SCORE</u>


```python
plt.figure(figsize=(12, 8))
sns.violinplot(x='Department', y='Absences', hue='PerformanceScore', data=df2, palette='husl')
plt.title('Absences Distribution by Department and Performance Score')
plt.xlabel('Department')
plt.ylabel('Absences')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.legend(title='Performance Score')
plt.show()
```

* INFERENCE : The distribution of absences varies across different departments.
Within each department, there are differences in absence patterns based on performance score. For example, in some departments, employees with higher performance scores tend to have lower absences. 

<u>SWARMPLOT OF SALARY DISTRIBUTION BY EMPLOYMENT STATUS</u>


```python
plt.figure(figsize=(10, 6))
sns.swarmplot(x='EmploymentStatus', y='Salary', data=df2, palette='viridis')
plt.title('Salary Distribution by Employment Status')
plt.xlabel('Employment Status')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

* INFERENCE : It provides a visual comparison of salary distributions across various employment statuses.
It may reveal any potential differences or similarities in salary distribution between different employment statuses.

<u>BOXENPLOT OF ABSENCES DISTRIBUTION BY GENDER AND MARITAL STATUS</u>


```python
plt.figure(figsize=(12, 8))
sns.boxenplot(x='Sex', y='Absences', hue='MaritalStatusID', data=df2, palette='Pastel1')
plt.title('Absences Distribution by Gender and Marital Status')
plt.xlabel('Gender')
plt.ylabel('Absences')
plt.grid(True)
plt.legend(title='Marital Status')
plt.show()
```

* INFERENCE : It allows us to compare the distribution of absences between different gender groups and marital status categories.
It can help identify any differences in absence patterns based on gender and marital status.

<u>HEATMAP OF DATAFRAME CORRELATION </u>


```python
numeric_df = df2.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

* INFERENCE : It helps identify the strength and direction of relationships between pairs of variables.
Positive values indicate a positive correlation (variables move in the same direction), negative values indicate a negative correlation (variables move in opposite directions), and values close to zero indicate weak or no correlation.
It can be used to identify potential multicollinearity issues or variables strongly correlated with the target variable.
